```python
from pathlib import Path
import pandas as pd
```

# Extract

```python
BRONZE = Path("data/bronze")
SILVER = Path("data/silver")
SILVER.mkdir(parents=True, exist_ok=True)

subs = pd.read_csv(
    BRONZE / "ravenstack_subscriptions.csv",
    parse_dates=["start_date","end_date"],
    dtype={"subscription_id": str, "account_id": str}
)

usage_raw = pd.read_csv(
    BRONZE / "ravenstack_feature_usage.csv",
    parse_dates=["usage_date"],
    dtype={"subscription_id": str, "usage_id": str}
)
```

# Transform

```python
# Trim IDs
subs["subscription_id"] = subs["subscription_id"].astype(str).str.strip()
usage_raw["subscription_id"] = usage_raw["subscription_id"].astype(str).str.strip()

# Join subscription dates
u = usage_raw.merge(
    subs[["subscription_id","start_date","end_date"]],
    on="subscription_id",
    how="left",
    validate="many_to_one"
).rename(columns={"start_date":"subscription_start_date",
                  "end_date":"subscription_end_date"})

# Keep only rows within the subscription window:
#   usage_date >= subscription_start_date
#   and if subscription_end_date is not null, usage_date <= subscription_end_date
mask_start_ok = u["usage_date"] >= u["subscription_start_date"]
mask_end_ok = u["subscription_end_date"].isna() | (u["usage_date"] <= u["subscription_end_date"])
before = len(u)
u_silver = u[mask_start_ok & mask_end_ok].copy()
dropped = before - len(u_silver)

print(f"[OK] Silver usage built. Kept {len(u_silver):,} rows, dropped {dropped:,} outside sub window.")
```

# Load

```python
# Save as silver
u_silver.to_csv(SILVER / "silver_feature_usage.csv", index=False)
```
